```python
import requests

def fetch_yahoo_news(api_key, query):
    base_url = "https://newsapi.org/v2/everything"
    params = {
        "q": query,
        "apiKey": api_key
    }

    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        articles = data.get("articles", [])
        for article in articles:
            print(f"Title: {article['title']}")
            print(f"Source: {article['source']['name']}")
            print(f"Description: {article['description']}")
            print(f"URL: {article['url']}")
            print("-" * 50)
    else:
        print("Failed to fetch news:", response.text)

# Replace 'YOUR_API_KEY' with your actual Yahoo API key
api_key = "d5c56486193b442b9c7d4f0017c420d7"
query = "Python programming"  # You can replace this query with your desired search query
fetch_yahoo_news(api_key, query)
```
